```python
import os
import sys
import logging
import datasets
import evaluate

import pandas as pd
import numpy as np

from transformers import DebertaV2ForSequenceClassification, DebertaV2Tokenizer, DataCollatorWithPadding
from transformers import Trainer, TrainingArguments
from peft import LoraConfig, get_peft_model, TaskType
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("/kaggle/working/labeledTrainData.tsv", header=0, delimiter="\t", quoting=3)
test = pd.read_csv("/kaggle/working/testData.tsv", header=0, delimiter="\t", quoting=3)
```

è®¾ç½®æ—¥å¿—è®°å½•çš„åŸºæœ¬é…ç½®ï¼Œå¹¶è®°å½•ç¨‹åºè¿è¡Œçš„å¼€å§‹

```python
program = os.path.basename(sys.argv[0])
logger = logging.getLogger(program)

logging.basicConfig(format='%(asctime)s: %(levelname)s: %(message)s')
logging.root.setLevel(level=logging.INFO)
logger.info(r"running %s" % ''.join(sys.argv))
```

å°†è®­ç»ƒæ•°æ®é›†åˆ†å‰²ä¸ºè®­ç»ƒé›†å’ŒéªŒè¯é›†ï¼Œå…¶ä¸­20%ä½œä¸ºéªŒè¯é›†

```python
train, val = train_test_split(train, test_size=.2)
```

åˆ›å»ºå­—å…¸ï¼Œå°†æ ‡ç­¾å’Œæ–‡æœ¬åˆ†åˆ«å­˜å‚¨ï¼Œä»¥é€‚é…æ¨¡å‹çš„è¾“å…¥æ ¼å¼ï¼Œå¹¶å°†å­—å…¸è½¬æ¢ä¸ºdatasetsåº“çš„æ•°æ®é›†å¯¹è±¡

```python
train_dict = {'label': train["sentiment"], 'text': train['review']}
val_dict = {'label': val["sentiment"], 'text': val['review']}
test_dict = {"text": test['review']}

train_dataset = datasets.Dataset.from_dict(train_dict)
val_dataset = datasets.Dataset.from_dict(val_dict)
test_dataset = datasets.Dataset.from_dict(test_dict)
```

æŒ‡å®šäº†é¢„è®­ç»ƒæ¨¡å‹çš„IDå¹¶ä»æŒ‡å®šçš„æ¨¡å‹IDåŠ è½½åˆ†è¯å™¨ã€‚

```python
model_id = "microsoft/deberta-v3-large"

tokenizer = DebertaV2Tokenizer.from_pretrained(model_id)
```

å®šä¹‰é¢„å¤„ç†å‡½æ•°ï¼Œå¯¹æ–‡æœ¬è¿›è¡Œåˆ†è¯å¤„ç†ï¼Œè®¾ç½®æˆªæ–­å’Œå¡«å……ï¼Œå¹¶å¯¹è®­ç»ƒé›†ã€éªŒè¯é›†å’Œæµ‹è¯•é›†åº”ç”¨é¢„å¤„ç†å‡½æ•°ã€‚

```python
def preprocess_function(examples):
    return tokenizer(examples['text'], truncation=True,padding='max_length', max_length=510)

tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_val = val_dataset.map(preprocess_function, batched=True)
tokenized_test = test_dataset.map(preprocess_function, batched=True)
```

åˆ›å»ºä¸€ä¸ªæ•°æ®å¡«å……å™¨ï¼Œç”¨äºåœ¨è®­ç»ƒæ—¶è‡ªåŠ¨å¡«å……æ‰¹æ¬¡ã€‚å¹¶ä»æŒ‡å®šçš„æ¨¡å‹IDåŠ è½½é¢„è®­ç»ƒæ¨¡å‹

```python
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
model = DebertaV2ForSequenceClassification.from_pretrained(
    model_id,
    # device_map="auto",
    # load_in_8bit=True
)
```

å®šä¹‰LoRAé…ç½®ï¼Œå¹¶ä½¿ç”¨peftåº“çš„å‡½æ•°å‡†å¤‡æ¨¡å‹ï¼Œæ‰“å°å¯è®­ç»ƒå‚æ•°ã€‚

```python
# Define LoRA Config
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    #target_modules=['q_proj', 'v_proj'],
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.SEQ_CLS
)
# prepare int-8 model for training
# model = prepare_model_for_kbit_training(model)

# add LoRA adaptor
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

åŠ è½½å‡†ç¡®ç‡ä½œä¸ºè¯„ä¼°æŒ‡æ ‡ï¼Œå¹¶å®šä¹‰å‡½æ•°è®¡ç®—æ¨¡å‹é¢„æµ‹çš„å‡†ç¡®ç‡ã€‚

```python
metric = evaluate.load("accuracy")
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

å®šä¹‰äº†è®­ç»ƒå‚æ•°ï¼ŒåŒ…æ‹¬è¾“å‡ºç›®å½•ã€è®­ç»ƒè½®æ•°ã€æ‰¹æ¬¡å¤§å°ç­‰ã€‚

```python
training_args = TrainingArguments(
    output_dir='./checkpoint',  # output directory
    num_train_epochs=3,  # total number of training epochs
    per_device_train_batch_size=2,  # batch size per device during training
    per_device_eval_batch_size=4,  # batch size for evaluation
    warmup_steps=500,  # number of warmup steps for learning rate scheduler
    weight_decay=0.01,  # strength of weight decay
    logging_dir='./logs',  # directory for storing logs
    logging_steps=100,
    save_strategy="no",
    evaluation_strategy="epoch"
)
```

åˆ›å»ºä¸€ä¸ªè®­ç»ƒå™¨å¯¹è±¡ï¼Œä¼ å…¥æ¨¡å‹ã€è®­ç»ƒå‚æ•°ã€æ•°æ®é›†ç­‰ã€‚

```python
trainer = Trainer(
    model=model,  # the instantiated ğŸ¤— Transformers model to be trained
    args=training_args,  # training arguments, defined above
    train_dataset=tokenized_train,  # training dataset
    eval_dataset=tokenized_val,  # evaluation dataset
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

å¼€å§‹è®­ç»ƒæ¨¡å‹ã€‚

```python
prediction_outputs = trainer.predict(tokenized_test)
test_pred = np.argmax(prediction_outputs[0], axis=-1).flatten()
print(test_pred)
```

```python
result_output = pd.DataFrame(data={"id": test["id"], "sentiment": test_pred})
result_output.to_csv("deb0erta_lora_int8.csv", index=False, quoting=3)
logging.info('result saved!')
```
